```python
import numpy as np
import pandas as pd
df = pd.read_csv("student or staff profiles.csv")
df
```

## df is ORIGINAL UNTOUCHED DATAFRAME ##

```python
ndf = df
ndf['person_id']=ndf['student_id'].fillna(df['staff_id'])
ndf = ndf[['entity_id','name','role','email','department','person_id','card_id','device_hash','face_id']]
ndf['has_face_id'] = ndf['face_id'].notna()
ndf
```

## ndf is the DATAFRAME WITH SINGLE COLUMN FOR STUDENT/STAFF_ID AND IT HAS A SEPARATE COLUMN INDICATING WHETHER FACE_ID IS PRESENT OR NOT ##

```python
nndf = ndf[ndf['has_face_id'] == False]
nndf.head()
```
