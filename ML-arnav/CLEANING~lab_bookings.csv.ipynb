```python
import numpy as np
import pandas as pd
df = pd.read_csv("lab_bookings.csv")
df
```

## df is the ORIGINAL UNTOUCHED DATAFRAME ##

```python
ndf = df
ndf['start_time']=pd.to_datetime(ndf['start_time'])
ndf['end_time']=pd.to_datetime(ndf['end_time'])
ndf['month'] = ndf['start_time'].dt.month
ndf['month_name'] = ndf['start_time'].dt.month_name()
ndf['day_of_week'] = ndf['start_time'].dt.dayofweek
ndf['day_name'] = ndf['start_time'].dt.day_name()
ndf['start_hour'] = ndf['start_time'].dt.hour
ndf['start_minute'] = ndf['start_time'].dt.minute
ndf['end_hour'] = ndf['end_time'].dt.hour
ndf['end_minute'] = ndf['end_time'].dt.minute
ndf
```

## ndf is the MODIFIED DATAFRAME WITH PROPER DATETIME ALIGNMENT ##
