```python
import numpy as np
import pandas as pd
df = pd.read_csv("campus card_swipes.csv")
df
```

## df is the ORIGINAL UNTOUCHED DATAFRAME ##

```python
ndf = df
ndf['timestamp']=pd.to_datetime(ndf['timestamp'])
ndf['month'] = ndf['timestamp'].dt.month
ndf['month_name'] = ndf['timestamp'].dt.month_name()
ndf['day_of_week'] = ndf['timestamp'].dt.dayofweek
ndf['day_name'] = ndf['timestamp'].dt.day_name()
ndf['hour'] = ndf['timestamp'].dt.hour
ndf['minute'] = ndf['timestamp'].dt.minute
ndf['second'] = ndf['timestamp'].dt.second
ndf
```

## ndf is the MODIFIED DATAFRAME WITH PROPER DATETIME ALIGNMENT ##
